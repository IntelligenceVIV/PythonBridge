```python
# How to create a Provider(供应者)
from faker import Faker
fake = Faker()
from faker.providers import BaseProvider
class Provider(BaseProvider):
    def foo(self):
        return 'bar'
fake.add_provider(Provider)
fake.foo()
```

```python
# How to customize the Lorem Provider
from faker import Faker
fake = Faker()
my_word_list = ['danish', 'cheesecake','suger','Lollipop','wafer','Gummies','sesame','Jelly','beans','pie','bar','Ice','oat']
fake.sentence()
fake.sentence(ext_word_list=my_word_list)
```
